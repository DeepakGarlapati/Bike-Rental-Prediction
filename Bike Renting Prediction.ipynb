## Exploring Data

### Visualizations - Univariate and Bivariate Analysis

```python
## We can see instant is just a serial number and has no relevant information
## As there are month and year feature extracted already, so dteday doesn't give much signifance as a independent variable
## As registerd + casual = count, we have count already, So the features registered and casual doesn't have individual significant value
## So these features can be dropped form our data

Bike_rent_data = Bike_rent_data.drop(["instant","dteday","casual","registered"], axis=1)
Bike_rent_data.head(10)    
```

```python
cat_var = ["season","weekday","workingday","holiday","yr","mnth","weathersit"]
num_var = ["cnt","windspeed","hum","temp","atemp"]
```

```python
%matplotlib inline

for i in cat_var:
    plt.figure()
    sns.countplot(Bike_rent_data[i]).set_title("Count plot of " +i )
    plt.figure()
    sns.barplot(Bike_rent_data[i],Bike_rent_data["cnt"]).set_title("Barplot of "+i+" vs bike count")
    
```

```python
%matplotlib inline

for i in num_var:
    plt.figure()
    sns.distplot(Bike_rent_data[i]).set_title("Histogram of " + i )
    plt.figure()
    sns.regplot(Bike_rent_data[i],Bike_rent_data["cnt"]).set_title("Scatterplot of " + i + " vs Bike rent count")
```

## Outlier Analysis

```python
for i in num_var:
    
    #Calculating 25th quartile and 75th quartile
    q75,q25 = np.percentile(Bike_rent_data.loc[:,i],[75,25])
    iqr = q75 - q25
    
    min = q25 - (1.5*iqr)
    max = q75 + (1.5*iqr)
    
    plt.figure()
    sns.boxplot(Bike_rent_data[i]).set_title("Distribution of " + i)
    
    Bike_rent_data.loc[Bike_rent_data.loc[:,i]<min,i] = min
    Bike_rent_data.loc[Bike_rent_data.loc[:,i]>max,i] = max
    
    plt.figure()
    sns.boxplot(Bike_rent_data[i]).set_title("Distribution of " + i + " after handling outliers")
    
```

## Feature scaling - Normalisation 

```python
#Normalising the variable 'cnt'
Bike_rent_data["cnt"] = (Bike_rent_data["cnt"]-np.min(Bike_rent_data["cnt"]))/(np.max(Bike_rent_data["cnt"])-np.min(Bike_rent_data["cnt"]))
```

```python
#Checking correlation using heat map
sns.heatmap(Bike_rent_data[num_var].corr())
Bike_rent_data[num_var].corr()
## temp and atemp highly correlated
```

```python
#Checking Multi Collinearity
MC = Bike_rent_data[num_var].assign(consnt =1)
pd.Series([variance_inflation_factor(MC.values,i) for i in range(MC.shape[1])],index = MC.columns)
## temp and atemp shows multicollinearity
```

```python
## ANOVA test for Categorical variables
for i in cat_var:
    print(i)
    print(stats.f_oneway(Bike_rent_data[i],Bike_rent_data['cnt']))
```

```python
# From the visualisations holiday and weekday doesn't give extra significance compared to working day, so both can be dropped
# temp and atemp are highly correlation, so either of them can be dropped
# As we are considering 95% success rate, we drop variable which have p-valve more than 0.05
Bike_rent_data = Bike_rent_data.drop(["holiday","workingday","atemp","yr"], axis = 1)
```

## Model Development

```python
#Sampling
#Splitting our data into train and test data with 75% data in train set
random.seed(102)
train_data, test_data = train_test_split(Bike_rent_data, test_size =  0.25)

x_train = train_data.drop("cnt",axis = 1)
y_train = train_data["cnt"]
x_test = test_data.drop("cnt",axis = 1 )
y_test = test_data["cnt"]
```

```python
#Checking the divided data sets
print(train_data.shape,test_data.shape)
test_data.head(5)
```

### Linear Regression

```python
# Creating a Linear regression model
from sklearn.linear_model import LinearRegression
Model_LR = sm.OLS(y_train,x_train).fit()
```

```python
#Summary of the the model
Model_LR.summary()
```

```python
#Predictting cnt using the model
predictions_LR = Model_LR.predict(x_test)
predictions_LR.head()
```

```python
# Calculate RMSE
math.sqrt(mean_squared_error(y_test,predictions_LR))
```

```python
#Calculate R-square
r2_score(y_test, predictions_LR)
```

```python
#Fucntion for Mean Absolute PErcentage Error
def MAPE(y_true, y_pred): 
    mape = np.mean(np.abs((y_true - y_pred) / y_true))*100 
    return mape
```

```python
#Calculating MAPE
MAPE(y_test, predictions_LR)
#Error = 33.530177840934996
#Accuracy = 66.469822159065
```

```python
#Plotting Actual vs Predicted values of Linear Regression model
x = list(range(len(test_data.index)))

plt.plot(x,y_test,x,predictions_LR)
plt.xlabel("x")
plt.ylabel("Values")
plt.legend(("Actual","Predicted"))
```

### Decision trees

```python
#Importing DecsionTreeRegressor from sklean.tree
from sklearn.tree import DecisionTreeRegressor
```

```python
#Building decision Tree Regression Model
Model_DT = DecisionTreeRegressor(max_depth = 6 ,max_features = 'auto')
```

```python
#Training the model
Model_DT.fit(x_train,y_train)
```

```python
#Predicting cnt the Decision Tree model
Predictions_DT = Model_DT.predict(x_test)
```

```python
#Calculating Rsquare and RMSE
print("R2:", Model_DT.score(x_test,y_test),"    RMSE:",math.sqrt(mean_squared_error(y_test,Predictions_DT)))
```

```python
#Calculating MAPE
MAPE(y_test, Predictions_DT)
#Error = 32.50752012567859
#Accuracy = 67.49247987432142
#If max_depth = 6
```

```python
#Plotting Actual vs Predicted values of Decision Tree Model
plt.plot(x,y_test,x,Predictions_DT)
plt.xlabel("x")
plt.ylabel("Values")
plt.legend(("Actual","Predicted"))
```
